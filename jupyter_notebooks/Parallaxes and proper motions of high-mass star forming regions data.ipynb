## Use astropy

```python
from astropy.io import ascii
from astropy import units as u
from astropy.coordinates import SkyCoord
```

```python
tbl = ascii.read('../data/asu.tsv', data_start=2)
tbl
```

```python
from astropy.coordinates import Longitude, Latitude

glon = Longitude(tbl['_Glon'] * u.deg)
glat = Latitude(tbl['_Glat'] * u.deg)
plt.scatter(glon, glat)
plt.xlabel('Galactic Longitude')
plt.ylabel('Galactic Latitude')
plt.show()
```

```python
glon = glon.wrap_at(180*u.degree)

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='mollweide')
ax.scatter(glon.radian, glat.radian)
ax.set_xticklabels(['14h','16h','18h','20h','22h','0h','2h','4h','6h','8h','10h'])
ax.grid(True)
plt.show()
```

```python
pd.unique(tbl['Arm'])
```

```python
arm_ids = pd.unique(tbl['Arm'])
listof_hmsfr = []

for k in range(len(arm_ids)):
    selector = (tbl['Arm'] == arm_ids[k])
    listof_hmsfr.append(SkyCoord(ra=tbl['RAJ2000'][selector], dec=tbl['DEJ2000'][selector], distance=1/tbl['plx'][selector]
             ,unit=['hourangle', 'deg', 'kpc'], equinox='J2000.0'))
```

```python
listof_hmsfr
```

```python
# Convert to galactocentric coordinates
for k in range(len(listof_hmsfr)):
    listof_hmsfr[k] = listof_hmsfr[k].galactocentric
```

```python
listof_hmsfr
```

```python
def plt_hmsfr(fig, ax):
    markers = ['^', 'o', 'd', 'X', '*', 'p', 'h', 's', 'H']

    for k in range(len(listof_hmsfr)):
        ax.scatter(-listof_hmsfr[k].x, listof_hmsfr[k].y, marker=markers[k], label=arm_ids[k], zorder=1)

    ax.scatter(0, 0, marker=(6, 2, 0), color='r', s=400)  # center of Milky Way
    ax.scatter(8.34, 0, marker='o', color='y', s=200)  # the sun

    ax.set_xlim((-5, 17))
    ax.set_ylim((-5, 17))
    x0,x1 = ax.get_xlim()
    y0,y1 = ax.get_ylim()
    ax.set_aspect(abs(x1-x0)/abs(y1-y0))
    #ax.grid(b=True, which='major', color='k', linestyle='--')
    plt.legend()
    plt.xlabel('x (kpc)')
    plt.ylabel('y (kpc)')
```

```python
fig, ax = plt.subplots(figsize=(10,10))

plt_hmsfr(fig, ax)
plt.show()
```

```python
from galpy.potential import evaluatePotentials, evaluateDensities, plotPotentials, MWPotential2014, SpiralArmsPotential
from galpy.util import bovy_coords
from astropy import units as u
```

```python
def plt_spirals(fig, ax, pot, x, y, extent):
    n = len(x)
    X, Y = np.meshgrid(x, y)
    potxy = np.zeros((n, n))

    for ii in range(n):
        for jj in range(n):
            R, phi, z = bovy_coords.rect_to_cyl(x[ii].value, y[jj].value, 0)
            R, z = (R, z) * u.kpc
            potxy[ii, jj] = evaluateDensities(pot, R, z, phi)

    c = ax.contourf(np.log(potxy.T), origin='lower', extent=extent, ncontours=21)
    ax.set_xlabel('x (kpc)')
    ax.set_ylabel('y (kpc)')
    ax.set_title('Potential')
    plt.colorbar(c)
    plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots(figsize=(10,10))

sp = SpiralArmsPotential(N=4, alpha=18*u.deg, Rs=9*u.kpc, r_ref=8*u.kpc, H=0.18*u.kpc)
pot = sp

x = np.linspace(-5, 17, 100) * u.kpc
y = np.linspace(-5, 17, 100) * u.kpc
extent = [-5, 17, -5, 17]
plt_spirals(fig, ax, pot, x, y, extent)
plt_hmsfr(fig, ax)
plt.show()
```
