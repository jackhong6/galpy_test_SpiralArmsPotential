# Exploring effects of a spiral arm perturbation on stellar motions

```python
import numpy as np
import matplotlib.pyplot as plt
from galpy.potential import SpiralArmsPotential, MWPotential2014, vcirc, evaluatePotentials, evaluateDensities, evaluateRforces
from astropy import units as u
```

```python
#sp = [SpiralArmsPotential(N=2, alpha=12.8*u.deg, Rs=3*u.kpc, r_ref=8.34*u.kpc, phi_ref=8.9*u.deg, H=0.18*u.kpc, Cs=[8./3./np.pi, 0.5, 8./15./np.pi])] \
#     + MWPotential2014
sp = [SpiralArmsPotential(N=2, alpha=12.8*u.deg, Rs=3*u.kpc, r_ref=8.34*u.kpc, phi_ref=8.9*u.deg, H=0.18*u.kpc)] \
     + MWPotential2014
v = []
p = []
d = []
r = []
phis = np.linspace(0, 2*np.pi, 1000)
for phi in phis:
    v.append(vcirc(sp, 8*u.kpc, phi=phi))
    p.append(evaluatePotentials(sp, 8*u.kpc, 0, phi))
    d.append(evaluateDensities(sp, 8*u.kpc, 0, phi))
    r.append(evaluateRforces(sp, 8*u.kpc, 0, phi))
```

```python
plt.plot(phis, v)
plt.show()

plt.plot(phis, p)
plt.show()

plt.plot(phis, d)
plt.show()

plt.plot(phis, r)
plt.show()
```
